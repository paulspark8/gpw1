```python
import string
import swifter
```

```python
from pandas import Series


def get_subject(text):
    if isinstance(text, str):
        text = fc.email.header.decode_header(text)[0][0]
    if isinstance(text, Series):
        text = text.str
    tmp = str(text).strip()
    return tmp.translate(str.maketrans('', '', string.punctuation))

def get_body_text(text):
    tmp = ''
    if isinstance(text, Series):
        tmp = text.str.strip()
    else: tmp = str(text)
    tmp = fc.html_text_extractor(tmp)
    tmp = fc.re.sub(r"( +)|[\n\r\t]+", " ", tmp)
    tmp = tmp.translate(str.maketrans('', '', string.punctuation))
    return tmp
```

```python
df['subject_body'] = df['subject'].swifter.apply(get_subject) +'\n'+df['body'].swifter.apply(get_body_text)
df
```

```python
df.drop_duplicates(subset='subject_body', inplace=True)
```

```python
df.drop(['subject', 'body'], axis=1, inplace=True)
df
```

```python
from nltk.tokenize import word_tokenize

df['tok'] = df['subject_body'].swifter.apply(word_tokenize)
```

```python
from nltk.stem import WordNetLemmatizer

wnl = WordNetLemmatizer()

df['lemmized'] = df['tok'].swifter.apply(lambda texte: ' '.join([wnl.lemmatize(words) for words in texte]))
```
